```python
# Test Null no Association between weathersit and season
pd.crosstab(df.weathersit,df.season)
```

```python
# Paired sample or Related Sample T test - Before & After
# BP of patient before taking tablet and BP of same patient after taking tablet

# Null - There is no significant difference in Average variable1 before & after event

# Alt - There is significant difference in Average variable1 before & after event
```

```python
# Test Null Average inflation before and after demonitization equal 12 months 

beforedemon = [3.35, 4.14, 5.30, 6.46, 6.13, 6.59, 5.86, 5.51, 5.53, 5.91, 6.32, 6.72 ]
afterdemon =  [2.59, 2.23, 1.86, 2.62, 2.61, 2.21, 1.09, 1.08, 1.79, 2.52, 2.89, 3.24]

print('Before Demonitization:', np.mean(beforedemon))
print('After Demonitization:' , np.mean(afterdemon))
```

```python
# Null - There is no significant difference in Average variable1 before & after demonitizatoin
# Alt  - There is significant difference in Average variable1 before & after event demonitization
```

```python
from scipy.stats import ttest_rel
```

```python
ttest_rel(beforedemon,afterdemon)
# Since pvalue is 3.600767294619255e-07 is less than 0.05, Rejct Null
```

```python
# Test Null Average inflation before and after GST equal 12 months 

beforegst = [1.08,1.09,2.21,2.61,2.62,1.86,2.23,2.59,3.35,4.14,5.30,6.46]
aftergst =  [1.79,2.52,2.89,3.24,3.97,4.00,5.11,4.74,4.36,3.97,3.96,3.93]

print('Before Demonitization:', np.mean(beforegst))
print('After Demonitization:' , np.mean(aftergst))
```

```python
ttest_rel(beforegst,aftergst)
# Since pvalue = 0.11695042528410868 is greater than 0.05, Fail to Reject Null
```

```python
ttest_ind(yyes.balance,yno.balance,equal_var=False)

# Since pvalue is 2.5211136918751468e-29 less than 0.05, REJECT NULL
```
